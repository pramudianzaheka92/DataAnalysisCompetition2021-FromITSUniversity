## 4. Modelling With Holtwinters

```python
# Melakukan pembagian menjadi data train dan data test
semifinal_data_train = semifinal_data[:4000] # Ambil data dari index 0 - 4000
semifinal_data_test = semifinal_data[1467:] # Ambil data dari index 4001 - 5467
```

```python
semifinal_data_train.head()
```

```python
semifinal_data_test.head()
```

```python
# Calling the model: holtwinters
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Memanggil fungsi ExponentialSmoothing yang telah di import
model_holtwinters = ExponentialSmoothing(semifinal_data_train.price_share_usd, 
                                         trend="add", seasonal="mul", seasonal_periods=6).fit()
```

Keterangan:
1. ExponentialSmoothing merupakan salah satu teknik peramalan yang melakukan penimbangan-penimbangan terhadap data masa lalu dengan cara eksponensial sehingga data paling akhir mempunyai bobot atau timbangan lebih besar dalam rata-rata bergerak.
2. train merupakan data train.
3. Pada nilai seasonal saya memakai nilai dari multiplicative karena dalam setiap pengamatannya selalu dilakukan dengan perkalian dari sebuah komponen yang telah ada. Rumus multiplicative adalah: y = base * trend * seasonality * residual.
4. Durasi/periode nilai seasonal/musiman yang saya gunakan adalah 4.

```python
prediction = model_holtwinters.forecast(5)
prediction
```

```python
# Harga sahan setelah dilakukan pemodelan
plt.figure(figsize = (8, 5))
semifinal_data_train["price_share_usd"].plot(legend=True, label="Train")
semifinal_data_test["price_share_usd"].plot(legend=True, label="Actual")

prediction.plot(legend=True, label="Forecast")

plt.title("Prediction of Saham Price in BBN Company With Holtwinters Model")
plt.xlabel("Value")
plt.ylabel("Saham Price")
plt.show()
```

```python
# Let's visually see the results
plt.figure(figsize=(13, 8))

plt.scatter(x=semifinal_data_train.index, y=semifinal_data_train["price_share_usd"], label="Training Data", color="yellow")
plt.scatter(x=semifinal_data_test.index, y=semifinal_data_test["price_share_usd"], label="Test Actuals", color="green")
plt.plot(prediction, label="Baseline", color="blue")

# Plot Labels, Legends etc
plt.xlabel("contributors_all_time")
plt.ylabel("price_share_usd")
plt.legend(loc='best')
plt.title("Stock Price Predictions Based on the Column of Code Analysis Contributors Every Time With Holt Winters Model")

plt.show()
```
